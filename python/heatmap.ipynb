```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel(r"D:\DATA ANALYTICS\Python\dataset\heatmap_data.xlsx")
df = pd.DataFrame(data)
df
```

```python
columns = ["Task Completion Rates", "Decision-Making Skills Rating", "Sales Revenue Generated"]
correlation_matrix = df[columns].corr().round(3)
df
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".3f")
plt.title("Correlation Heatmap of Key Metrics")
plt.show()
```
